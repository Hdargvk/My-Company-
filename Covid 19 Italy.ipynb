# Grouping And aggregation 

```python
_df
```

```python
_month_df= _df.groupby('month')[['new_cases','new_deaths','new_tests']].sum()
```

```python
_month_df
```

```python
_month_df.loc[6]
```

```python
year_df= _df.groupby('year')[['new_cases','new_deaths','new_tests']].sum()
```

```python
year_df
```

```python
_df['total_cases']=_df.new_cases.cumsum()
```

```python
_df['total_deaths']=_df.new_deaths.cumsum()
```

```python
_df['total_tests']=_df.new_tests.cumsum()
```

```python
_df
```

# Writing data back to files
